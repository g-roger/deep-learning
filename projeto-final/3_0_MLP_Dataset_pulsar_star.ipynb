```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, plot_confusion_matrix,accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn import datasets
from keras.models import Sequential, Model
from keras.layers import Input, Dense
from keras.layers import Dropout
from sklearn import metrics

import pydot

import numpy as np
import seaborn as sns
```

```python
pulsar_df = pd.read_csv('https://raw.githubusercontent.com/g-roger/deep-learning/master/projeto-final/datasets/pulsar_data_train.csv')
```

```python
pulsar_df.astype({'target_class': 'int32'}).dtypes
```

```python
pulsar_df['target'] = pulsar_df['target_class'].apply(np.int64)
```

```python
pulsar_df
```

```python
pulsar_df.drop(columns=['target_class'], inplace=True)
pulsar_df.dropna()
```

```python
# Normalização

X=pulsar_df.drop(['target'],axis=1)
y=pulsar_df['target']
X=StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=25)
```

```python
# Criação de rede MLP com KERAS
#Modelo da Rede MLP

model = Sequential()
model.add(Dense(10, input_dim = X_train.shape[1], activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(8, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=80)
```

```python
scores = model.evaluate(X_test, y_test)
```
